```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df1=pd.read_csv('popularity.csv')
df1
```

```python
def removeOutliers(a,outlierConstant):
    
    uq=np.percentile(a,95)
    lq=np.percentile(a,10)
    iqr=(uq-lq)*outlierConstant
    q=(lq-iqr,uq+iqr)
    r=[]
    for y in a.tolist():
        if y>=q[0] and y<=q[1]:r.append(y)
    return r
removeOutliers(df1[' weekday_is_monday'],5)
```

```python
Q1 = df1.quantile(0.25)
Q3 = df1.quantile(0.75)
IQR = Q3 - Q1
((df1 < (Q1 - 1.5 * IQR)) | (df1 > (Q3 + 1.5 * IQR))).sum()
```

```python
def remove_outlier(df_in, col_name):
    q1=df_in[col_name].quantile(0.25)
    q3=df_in[col_name].quantile(0.75)
    iqr=q3-q1 #Interquartile range
    fence_low=q1-1.5*iqr
    fence_high=q3+1.5*iqr
    df_out=df_in[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    return df_out

re_dat = remove_outlier(df1,' shares')
re_dat
```

```python
sns.boxplot(re_dat[' shares'])
```
