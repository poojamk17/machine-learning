```python
import numpy as np
import pandas as pd
df=pd.read_csv('mt_cars.csv')
df
```

```python
df[df['cyl']==6][['disp']]
```

```python
a=df[df['cyl']==6][['disp']].mean()
a
```

```python
df[df['cyl']==6][['disp']].fillna(value=a)
```

```python
df[df['cyl']==6][['disp']].fillna(value=a,inplace=True)
df
```

```python
df['disp'].fillna(value=a,inplace=True)
```

```python
df
```

```python
a=df[df['cyl']==6]['disp'].mean()
df['disp']=df.apply(lambda row: a 
                    if(row.cyl==6) & (np.isnan(row['disp']))
                    else row.disp,axis=1)
df
```
