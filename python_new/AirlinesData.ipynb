```python
# airlines with fatalities increased in  (00-14 to 85-99)
df2 = df[df['fatal_accidents_00_14']>df['fatal_accidents_85_99']]
df2
```

```python
# print no of fatalities increased in new column
df2['fat_inc_8599_to_0014'] = df2.apply(lambda b: b.fatal_accidents_00_14-b.fatal_accidents_85_99 if b.fatal_accidents_00_14>b.fatal_accidents_85_99 else b.fatal_accidents_85_99-b.fatal_accidents_00_14,axis=1)
df2
```

```python
# display fatalities % increased in new column
df2['Perc_inc'] = df2.apply(lambda b: (b.fatal_accidents_00_14-b.fatal_accidents_85_99)*100/(b.fatal_accidents_00_14) if b.fatal_accidents_00_14>b.fatal_accidents_85_99 else b.fatal_accidents_85_99-b.fatal_accidents_00_14,axis=1)
df2
```

```python
df2['Perc_inc']
```

```python
# Calculate percentage of fatalities Increased/Decreased in new column
df['fatalities_increased/decreased'] = df.apply(lambda b: 0 if b.fatal_accidents_00_14==b.fatal_accidents_85_99 else (b.fatal_accidents_00_14-b.fatal_accidents_85_99 if b.fatal_accidents_00_14>b.fatal_accidents_85_99 else b.fatal_accidents_00_14-b.fatal_accidents_85_99),axis=1)
df
```

```python
# print % of increased/decreased in fatalities in new column
df['Percentage Increased/Decreased'] = df.apply(lambda b: 0 if b.fatal_accidents_00_14==b.fatal_accidents_85_99 else ( (b.fatal_accidents_00_14-b.fatal_accidents_85_99)*100/(b.fatal_accidents_00_14) if b.fatal_accidents_00_14>b.fatal_accidents_85_99 else (b.fatal_accidents_00_14-b.fatal_accidents_85_99)*100/(b.fatal_accidents_85_99) ),axis=1)
df
```

```python
# print 'Not changed' for 0%
# print 'Increased' for fatatities% increased
# print 'Decreased' for fatalities% decreased in new column
df['status'] = df.apply(lambda c: 'Not Changed' if c.fatal_accidents_00_14==c.fatal_accidents_85_99 else ('Increased' if c.fatal_accidents_00_14>c.fatal_accidents_85_99 else 'Decreased'),axis=1)
df
```
