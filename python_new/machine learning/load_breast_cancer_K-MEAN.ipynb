```python
plt.scatter(df3_sc[:,0],df3_sc[:,1])
```

```python
df3_sc_df=pd.DataFrame(df3_sc,columns=df3.columns)
```

```python
X=df3_sc_df
```

```python
from sklearn.cluster import KMeans
cluster_id=[]
K=list(range(1,15))
for i in K:
    km=KMeans(i).fit(X)
    y_km_df_pred=km.fit_predict(X)
    cluster_id.append(km.inertia_)
cluster_id
```

```python
plt.plot(K,cluster_id)
```

```python
### fitting K-MEAN to the dataset
kmeans=KMeans(n_clusters=2,init='k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(X)
y_kmeans
```

```python
plt.scatter(df3_sc[y_kmeans==0,5],df3_sc[y_kmeans==0,8],s=50,c='red',label='cluster1')
plt.scatter(df3_sc[y_kmeans==1,10],df3_sc[y_kmeans==1,9],s=50,c='blue',label='cluster2')
# plt.scatter(df_sc[y_kmeans==2,3],df_sc[y_kmeans==2,2],s=50,c='green',label='cluster3')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='yellow',label='centroid')
plt.title('clusters of mall')
plt.legend()
plt.show()
```
