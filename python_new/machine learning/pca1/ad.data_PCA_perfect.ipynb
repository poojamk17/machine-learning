```python
df1=df[(df[' 125']>(np.mean(df[' 125'])-(3*np.std(df[' 125'])))) & (df[' 125']<(np.mean(df[' 125'])+3*np.std(df[' 125'])))]
df1.shape                                                                                           
```

```python
df2=df1[(df1['   1.0']>(np.mean(df1['   1.0'])-(3*np.std(df1['   1.0'])))) & (df1['   1.0']<(np.mean(df1['   1.0'])+3*np.std(df1['   1.0'])))]
df2.shape                                                                                           
```

```python
# df3=df2[(df2['   1.0']>(np.mean(df2['   1.0'])-(3*np.std(df2['   1.0'])))) & (df2['   1.0']<(np.mean(df2[' 125'])+2.5*np.std(df2['   1.0'])))]
# df3.shape                                                                                           
```

```python
df2.skew()
```

```python
X=df2.drop('ad.',axis=1)
y=df2['ad.']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
X_train[[' 125',' 125.1','   1.0']]=sc.fit_transform(X_train[[' 125',' 125.1','   1.0']])
```

```python
X_test[[' 125',' 125.1','   1.0']]=sc.transform(X_test[[' 125',' 125.1','   1.0']])
```

```python
X_test
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=50)
```

```python
X_train_sc_pc=pca.fit_transform(X_train)
X_test_sc_pc=pca.transform(X_test)
```

```python
X_train_sc_pc
```

```python
pca.explained_variance_ratio_
```

```python

x_pca_df_train=pd.DataFrame(X_train_sc_pc)
x_pca_df_test=pd.DataFrame(X_test_sc_pc)
```

```python
x_pca_df_train
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier() ### default K=8
knn.fit(X_train_sc_pc,y_train)
```

```python
# pd.set_option('display.max_columns',None)
# df3.head()
```

```python
pre_train=knn.predict(X_train_sc_pc)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print('accuracy_score')
print(accuracy_score(y_train,pre_train))
```

```python
print('\n\nclassification report: \n\n',classification_report(y_train,pre_train))
```

```python
pre_test=knn.predict(X_test_sc_pc)
```

```python
print('accuracy_score')
print(accuracy_score(y_test,pre_test))
```

```python
print('\n\nclassification report: \n\n',classification_report(y_test,pre_test))
```
