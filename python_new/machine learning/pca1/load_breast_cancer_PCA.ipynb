```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### DATA

```python
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
```

```python
cancer.keys()
```

```python
df=pd.DataFrame(cancer.data,columns=cancer.feature_names)
```

```python
df
```

### scale

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler() 
```

```python
df_sc=sc.fit_transform(df)
```

```python
df_sc
```

## PCA ANALYSIS

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=5)
```

```python
pca.fit(df_sc)
```

```python
x_pca=pca.transform(df_sc)
```

```python
df_sc.shape
```

```python
x_pca.shape
```

```python
x_pca
```

```python
x_pca[:,0] ## pc1
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'],cmap='plasma')
plt.xlabel('first principal component')
plt.ylabel('second principal component')
```

## interpreting the component

```python
pca.components_ ## components like 2,5,7,in example and from this we calculate P11,P21...
```

```python
df_comp=pd.DataFrame(pca.components_,columns=cancer['feature_names'],index=['PC1','PC2','PC3','PC4','PC5'])
```

```python
df_comp
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
pc=['PC1','PC2','PC3','PC4','PC5']
plt.bar(pc,pca.explained_variance_ratio_)
```

```python
x_pca_df=pd.DataFrame(x_pca,columns=pc)
```

```python
x_pca_df
```

### split into X and Y do analysis

```python
X=x_pca_df
y=cancer.target
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(2) ### default K=8
knn.fit(X_train,y_train)
```

```python
pre_train=knn.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print('accuracy_score')
print(accuracy_score(y_train,pre_train))
```

```python
pre_test=knn.predict(X_test)
```

```python
print('accuracy_score')
print(accuracy_score(y_test,pre_test))
```

### LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
```

```python
lgr=LogisticRegression()
```

```python
lgr.fit(X_train,y_train)
```

```python
y_train_pred=lgr.predict(X_train)
len(y_train_pred)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print('accuracy_score')
print(accuracy_score(y_train,y_train_pred.round()))
```

```python
print('confusion matrix\n\n',confusion_matrix(y_train,y_train_pred.round()))
```

```python
print('\n\nclassification report: \n\n',classification_report(y_train,y_train_pred.round()))
```
