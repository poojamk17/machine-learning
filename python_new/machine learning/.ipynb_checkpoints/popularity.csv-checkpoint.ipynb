```python
def remove_outlier(df_in, col_name):
    q1=df_in[col_name].quantile(0.25)
    q3=df_in[col_name].quantile(0.75)
    iqr=q3-q1 #Interquartile range
    fence_low=q1-1.5*iqr
    fence_high=q3+1.5*iqr
    df_out=df_in[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    return df_out

df1 = remove_outlier(df,' n_unique_tokens')
df1
```

```python
X=df1.drop([' shares','url'],axis=1)
y=df1[' shares']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
print(lm.intercept_)
```

```python
lm.coef_
```

```python
y_pred=lm.predict(X_test)
```

```python
y_pred_train=lm.predict(X_train)#check/predect for train data output 
```

```python
len(y_pred)
```

```python
len(y_test)
```

```python
from sklearn import metrics
```

```python
metrics.r2_score(y_test,y_pred)#91% data is correct
```

```python
metrics.r2_score(y_train,y_pred_train)
```
