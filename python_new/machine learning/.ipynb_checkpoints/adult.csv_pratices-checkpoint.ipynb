```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df1=pd.read_csv('adult.csv',na_values=['?'])
df1.head()
```

```python
df1.info()
```

```python
df1.isnull().sum()
```

```python
df1['workclass'].fillna(method='bfill',inplace=True)
```

```python
df1['occupation'].fillna(method='bfill',inplace=True)
```

```python
df1['native-country'].fillna(method='bfill',inplace=True)
```

```python
df1.isnull().sum()
```

```python
df1['workclass'].value_counts()
```

```python
status=pd.get_dummies(df1['workclass'],drop_first=True)
status.head()
```

```python
df1=pd.concat([df1,status],axis=1)
df1.drop('workclass',axis=1,inplace=True)
df1.head()
```

```python
status1=pd.get_dummies(df1['education'],drop_first=True)
status1.head()
```

```python
df1=pd.concat([df1,status1],axis=1)
df1.drop('education',axis=1,inplace=True)
df1.head()
```

```python
status2=pd.get_dummies(df1['marital-status'],drop_first=True)
df1=pd.concat([df1,status2],axis=1)
df1.drop('marital-status',axis=1,inplace=True)
df1.head()
```

```python
status2=pd.get_dummies(df1['occupation'],drop_first=True)
df1=pd.concat([df1,status2],axis=1)
df1.drop('occupation',axis=1,inplace=True)
df1.head()
```

```python
status2=pd.get_dummies(df1['relationship'],drop_first=True)
df1=pd.concat([df1,status2],axis=1)
df1.drop('relationship',axis=1,inplace=True)
df1.head()
```

```python
status2=pd.get_dummies(df1['race'],drop_first=True)
df1=pd.concat([df1,status2],axis=1)
df1.drop('race',axis=1,inplace=True)
df1.head()
```

```python
status2=pd.get_dummies(df1['gender'],drop_first=True)
df1=pd.concat([df1,status2],axis=1)
df1.drop('gender',axis=1,inplace=True)
df1.head()
```

```python
status2=pd.get_dummies(df1['native-country'],drop_first=True)
df1=pd.concat([df1,status2],axis=1)
df1.drop('native-country',axis=1,inplace=True)
df1.head()
```

```python
status2=pd.get_dummies(df1['income'],drop_first=True)
df1=pd.concat([df1,status2],axis=1)
df1.drop('income',axis=1,inplace=True)
df1.head()
```
