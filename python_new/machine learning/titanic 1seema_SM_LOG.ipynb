```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("titanic.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop('Cabin',axis=1,inplace=True)
```

```python
df.columns
```

```python
df.drop('PassengerId',axis=1,inplace=True)
```

```python
df.drop('Name',axis=1,inplace=True)
```

```python
df.drop('Ticket',axis=1,inplace=True)
```

```python
df['Age'].fillna(method='bfill',inplace=True)
```

```python
df['Embarked'].fillna(method='bfill',inplace=True)
```

```python
a=pd.get_dummies(df['Embarked'],drop_first=True)
df=pd.concat([df,a],axis=1)
a.head()
```

```python
df.isnull().sum()
```

```python
df.drop('Embarked',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df['Sex']=le.fit_transform(df['Sex'])
```

```python
df.skew()
```

```python
df.kurtosis()
```

```python
df1=df[(df['Fare']>(np.mean(df['Fare'])-3*np.std(df['Fare']))) & (df['Fare']<(np.mean(df['Fare'])+2*np.std(df['Fare'])))]
```

```python
df1.skew()
```

```python
df1.shape
```

```python
df1.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df1[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Q', 'S']]
y=df1['Survived']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,test_size=0.3,random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train[['Fare','Age']]=scaler.fit_transform(X_train[['Fare','Age']])
X_test[['Fare','Age']]=scaler.transform(X_test[['Fare','Age']])
X_train_scaled=X_train
X_test_scaled=X_test
X_train.head()
```

```python
type(X_test)
```

```python
type(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
lgr=LogisticRegression()
```

```python
from sklearn.feature_selection import RFE
rfe=RFE(lgr,5)
```

```python
rfe=rfe.fit(X_train_scaled,y_train)
```

```python
rfe.support_
```

```python
rfe.ranking_
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
selected_col=X_train_scaled.columns[rfe.support_]
```

```python
type(selected_col)
```

```python
selected_col
```

```python
X_train_scaled_rfe=X_train_scaled[selected_col]
type(X_train_scaled_rfe)
```

```python
lgr.fit(X_train_scaled_rfe,y_train)
```

```python
import statsmodels.api as sm
```

```python
X_train_scaled_rfe_sm=sm.add_constant(X_train_scaled_rfe)
```

```python
lgmod=sm.GLM(y_train,X_train_scaled_rfe_sm,family=sm.families.Binomial()).fit()
```

```python
print(lgmod.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif=pd.DataFrame()
vif['Features']=X_train_scaled_rfe_sm.columns
vif['VIF']=[variance_inflation_factor(X_train_scaled_rfe_sm.values,i) for i in range(X_train_scaled_rfe_sm.shape[1])]
vif["VIF"]=round(vif["VIF"],2)
vif=vif.sort_values(by="VIF",ascending=False)
vif
```

```python
y_pred_sm=lgmod.predict(X_train_scaled_rfe_sm)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print(' Accuracy Score:\n')
print(accuracy_score(y_train,y_pred_sm.round()))
```

```python
print("confusion Matrix:\n\n",confusion_matrix(y_train,y_pred_sm.round()))
```

```python
X_test_scaled_rfe=X_test[X_train_scaled_rfe.columns]
X_test_scaled_rfe.head()
```

```python
y_pred_lgr_test=lgr.predict(X_test_scaled_rfe)
```

```python
print(' Accuracy Score:\n')
print(accuracy_score(y_test,y_pred_lgr_test.round()))
```

```python
print("confusion Matrix:\n\n",confusion_matrix(y_test,y_pred_lgr_test.round()))
```

```python
print("\n\nClassification Report:\n",classification_report(y_test,y_pred_lgr_test))
```
