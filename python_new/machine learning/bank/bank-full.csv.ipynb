```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
X_train[['age','balance','day','campaign']]=sc.fit_transform(X_train[['age','balance','day','campaign']])
```

```python
X_test[['age','balance','day','campaign']]=sc.transform(X_test[['age','balance','day','campaign']])
```

```python
X_train.head()
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
lgr=LogisticRegression()
```

```python
rfe=RFE(lgr,10)
```

```python
rfe=rfe.fit(X_train,y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
selected_col=X_train.columns[rfe.support_]
selected_col
```

```python
X_train_sc_rfe=X_train[selected_col]
```

```python
X_train_sc_rfe.head()
```

```python
X_train_sc_rfe.shape
```

```python
lgr.fit(X_train_sc_rfe,y_train)
```

```python
y_train_sm_pred=lgr.predict(X_train_sc_rfe)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print('accuracy_score')
print(accuracy_score(y_train,y_train_sm_pred))
```

```python
print('confusion matrix\n\n',confusion_matrix(y_train,y_train_sm_pred))
```

```python
print('\n\nclassification report: \n\n',classification_report(y_train,y_train_sm_pred))
```

```python
X_test=X_test[X_train_sc_rfe.columns]
```

```python
y_test_sm_pred=lgr.predict(X_test)
```

```python
print('accuracy_score')
print(accuracy_score(y_test,y_test_sm_pred))
```

```python
print('\n\nclassification report: \n\n',classification_report(y_test,y_test_sm_pred))
```
