```python
#map--->list and apply---->dataframe
d={'cost':[5000,7000,9000,12000],
  'disc':[0.1,0.2,0.15,0.25]}
df=pd.DataFrame(d)
df
```

```python
df['cost with GST @ 18%']=df.apply(lambda c: c.cost+(c.cost*(18/100)), axis=1)
df
```

```python
df['cost with discount']=df.apply(lambda row: row.cost-(row.cost*row.disc), axis=1)
df
```
