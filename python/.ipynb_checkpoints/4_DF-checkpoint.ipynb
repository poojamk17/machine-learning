```python
a={'A':[1,2,3,4],
   'B':[5,6,7,8],
   'C':[9,10,1,2],
   'D':[2,4,8,6]}

b={'E':[10,20,30,40],
   'B':[50,60,70,80],
   'F':[90,100,10,20],
   'G':[20,40,8,60]}

c={'H':[100,200,300,400],
   'B':[500,600,700,800],
   'I':[900,1000,100,200],
   'J':[200,400,8,600]}

```

```python
df1=pd.DataFrame(a)
df2=pd.DataFrame(b,index=[4,5,6,7])
df3=pd.DataFrame(c,index=[8,9,10,11])
```

```python
pd.merge(df1,df2,on='B')
```

```python
pd.merge(df1,df2,on='B',how='outer')
```

```python
pd.merge(df1,df2,on='B',how='left')
```

```python
pd.merge(df1,df2,on='B',how='right')
```
