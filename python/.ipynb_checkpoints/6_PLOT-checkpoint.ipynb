```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[12,13,14,11]
```

```python
plt.plot(x,y,'r')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('graph')
plt.show()
```

```python
plt.scatter(x,y)
```

```python
plt.bar(x,y)
```

```python
plt.pie(x,y)
plt.show()
```

```python
plt.plot(x,y,kind='bar')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x,x)
```

```python
fig=plt.figure()
axis=fig.add_axes([1,1,1,1])
axis.plot(x,y)
axis.set_xticks([1,2,3,4,5,6,7,8,9])
axis.set_xticklabels(['a','b','c','d','e','f','g','h','i'],rotation=60,fontsize=25)
plt.show()
```

```python
fig=plt.figure()
axis=fig.add_axes([0.1,0.1,1,1])
axis.plot(x,y)
ax=fig.add_axes([0.2,0.5,0.4,0.3])
ax.plot(x,x)
```

```python
x=np.arange(10)
y=x**2
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,1,1])
axes1.plot(x,y)
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes2.plot(x,y)
axes1.set_xlabel('x axis')
axes1.set_ylabel('y axis')

```
